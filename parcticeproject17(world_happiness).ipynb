```python
import scipy
import scipy as stats
from scipy.stats import zscore
from scipy.stats import percentileofscore
z=abs(zscore(df))
```

```python
print(np.where(z>3))
```

```python
dff=df[(z<3).all(axis=1)]
```

```python
print("original shape=",df.shape,"\nNew shape=",dff.shape)
```

# Removing skewness

```python
dff.skew()
```

```python
x=dff.iloc[:,1:]
```

```python
y=dff.iloc[:,0]
```

```python
x.head()
```

```python
from sklearn.preprocessing import power_transform
x=power_transform(x,method=("yeo-johnson"))
```

```python
from sklearn.preprocessing import StandardScaler
sl=StandardScaler()
print(sl.fit(dff))
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.33, random_state=42)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lm=LinearRegression()
```

```python
lm.fit(x_train,y_train)
```

```python
#predict the value
pred=lm.predict(x_test)
print("predict result price",pred)
print("actual price",y_test)
```

```python
lm.coef_
```

```python
lm.intercept_
```

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error
print("error:")

print("mean absolute error",mean_absolute_error(y_test,pred))
print("mean squared error",mean_squared_error(y_test,pred))
print("root mean squared error",np.sqrt(mean_squared_error(y_test,pred)))
```

```python
lm.score(x_train,y_train)
```

```python
from sklearn.metrics import r2_score
print(r2_score(y_test,pred))
```

```python
from sklearn.ensemble import RandomForestRegressor
ran=RandomForestRegressor(n_estimators=200)
ran.fit(x_train,y_train)
pred=ran.predict(x_test)
```

```python
pred=ran.predict(x_train)
```

```python
r2_score=r2_score(y_test,pred)
r2_score
```

```python
from sklearn.svm import SVR
klist=["poly", 'linear',"rbf"]
for i in klist:
    sv=SVR(kernel=i)
    sv.fit(x_train,y_train)
    print(sv.score(x_train, y_train))
```

```python
pred=sv.predict(x_test)
result=pd.DataFrame({
    'Actual':y_test,
    'Predict':pred, "Difference":(y_test-pred)})
print(result)
sns.regplot(x='Actual',y='Predict',data=result)
```
